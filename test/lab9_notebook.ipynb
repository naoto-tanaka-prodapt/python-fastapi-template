```python
import pandas as pd
```

```python
csv = pd.read_csv("./job_rewriting_tests.csv")
display(csv)
```

```python
import sys
sys.path.append("..")
from llm.review_description import review_description, RevisedDescription

summaries = []
revised_descriptions = []

for row in csv.itertuples():
    reviewed: RevisedDescription = review_description(row[1])
    summaries.append(reviewed.overall_summary)
    revised_descriptions.append(reviewed.rewritten_description)

csv["Summary"] = summaries
csv["Fixed Description"] = revised_descriptions
```
